# 1. Tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

desc_stats = numeric_df.describe().T

desc_stats['Q1'] = numeric_df.quantile(0.25)
desc_stats['Q2'] = numeric_df.quantile(0.50)
desc_stats['Q3'] = numeric_df.quantile(0.75)

desc_stats[['mean','std','min','max','Q1','Q2','Q3']]
```

# 5. Visualisasikan dataset dengan salah satu grafik. Pilih grafik yang paling cocok untuk memvisualisasikan data tersebut.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Age', y='PlayTimeHours', hue='Gender', alpha=0.3, s=20)
plt.title("Hubungan Usia dan Lama Bermain Game")
plt.show()
```

# 6.Tampilkan korelasi dataset menggunakan heatmap.


```python
corr = numeric_df.corr()

plt.figure(figsize=(10,6))
plt.imshow(corr, cmap='coolwarm', interpolation='none')
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Heatmap Korelasi")
plt.show()
```

# Kesimpulan Analisis Dataset
Berdasarkan hasil analisis dan visualisasi data, diperoleh beberapa poin penting sebagai berikut:

1. **Statistik Deskriptif**
   - Data numerik menunjukkan distribusi yang bervariasi pada tiap atribut.
   - Informasi seperti rata-rata, nilai minimum, maksimum, serta kuartil membantu memahami penyebaran data.

2. **Korelasi Antar Atribut**
   - Heatmap korelasi memperlihatkan adanya hubungan yang cukup kuat antar beberapa atribut.
   - Korelasi positif berarti kenaikan satu variabel cenderung diikuti oleh kenaikan variabel lain.
   - Korelasi negatif berarti kenaikan satu variabel diikuti oleh penurunan variabel lain.

3. **Visualisasi Scatter Plot**
   - Scatter plot memperlihatkan pola distribusi data antar dua atribut.
   - Beberapa pasangan variabel menunjukkan pola linier, sementara yang lain menyebar acak tanpa pola yang jelas.

4. **Outlier**
   - Dari scatter plot terlihat adanya titik data yang jauh dari mayoritas (outlier).
   - Outlier ini dapat memengaruhi hasil analisis dan perlu diperhatikan lebih lanjut.

Dari hasil analisis ini, dapat disimpulkan bahwa data memiliki karakteristik yang beragam, terdapat hubungan antar variabel, serta perlu perhatian khusus terhadap outlier sebelum melanjutkan ke tahap analisis yang lebih mendalam.
